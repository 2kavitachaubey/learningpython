```python
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
iris.dtypes
```

```python
## correlation with data heatmap()
```

```python
iris.corr()
```

```python
sns.heatmap(iris.corr())
```

## distribution plot
### Jointplot
### pairplot
### distplot

```python
## Jointplot
```

```python
sns.jointplot(x='sepal_length',y='petal_length',data=iris,kind='hex')
```

```python
sns.jointplot(x='sepal_width',y='petal_width',data=iris,kind='reg')
```

```python
## pairplot
```

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris,hue='sepal_width')
```

```python
sns.pairplot(iris,hue='petal_length')
```

```python
iris['sepal_width'].value_counts()
```

```python
## Distplot
```

```python
sns.distplot(iris['sepal_length'])
```

```python
sns.distplot(iris['petal_width'],kde=False,bins=10)
```

## categorical Plots
### countplot
### barplot
### boxplot
### violinplot

```python
### countplot
```

```python
sns.countplot('sepal_width',data=iris)
```

```python
sns.countplot(y='petal_length',data=iris)
```

```python
### bar plot
```

```python
sns.barplot(x='sepal_length',y='petal_length',data=iris)
```

```python
sns.barplot(x='sepal_width',y='petal_width',data=iris)
```

```python
### boxplot
```

```python
sns.boxplot('sepal_length','sepal_width',data=iris)
```

```python
sns.boxplot('sepal_length','sepal_width',data=iris,palette='rainbow')
```

```python
sns.boxplot('petal_length','petal_width',data=iris,orient='h')
```

```python
### violinplot
```

```python
sns.violinplot(x='petal_length',y='petal_width',data=iris,palette='rainbow')
```

```python
sns.violinplot(x='petal_length',y='petal_width',data=iris,palette='rainbow',orient='h')
```
