# MatplitLib 
### Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy. It provide a object-oriented API for embedding plots into application using general-purpose GUI toolkits like Tkinter, wxPython, QT, or GTK+

## Some of the major Pros of MatplotLib are:
- Generally easy to get started for simple plots
- Support for custom labels and texts
- Great control of every element in a figure
- High-quality output in many formats
- Very customizable in general

```python
##plotting using matplotlib

##plt scatter

plt.scatter(x,y,c='g')   ## value of x,value of y,colour of dots
plt.xlabel('X axis')  ## name for x axis
plt.ylabel('Y axix')
plt.title('Graph in 2D')
```

```python
## more plot

plt.plot(x,y,'r*',linestyle='dashed',linewidth=2,markersize=8) ## x,y,colour,line style, line width, mark size
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Designed graph')
```

```python
## creating subplots

plt.subplot(2,2,1) ## rows, column, index -- 2 row,2 column, index 1
plt.plot(x,y,'r*-')
plt.subplot(2,2,2)
plt.plot(x,y,'b-')
plt.subplot(2,2,3)
plt.plot(x,y,'co')    
# not making the forth graph
```

### sine and cosine wave graph

```python
#subplot()
# compute the x and y coordinates for points on sine and cosine curves
x = np.arange(0,5*np.pi,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

# set the height of 2 and width of 1
# set the subplot for sin as active and plot it
plt.subplot(2,1,1)
plt.plot(x,y_sin)
plt.title("Sine")

# set the subplot for cos as active
plt.subplot(2,1,2)
plt.plot(x,y_cos)
plt.title("Cosine")
```

# Bar graph

```python
## Bar plot
x= [2,8,10]
y= [11,16,9]

x2= [3,9,11]
y2= [6,15,7]
plt.bar(x,y)
plt.bar(x2,y2,color = 'g')
plt.title("Bar graph")
plt.xlabel("x axis")
plt.ylabel("y axis")
```

# Histograms

```python
a = np.array([5,6,2,11,17,22,23,24,45,49,55,64,65,66,81])
plt.hist(a,bins=20)
plt.title("Histogram")
```

# Box Plot using Matplotlib

```python
data = [np.random.normal(0,std,100) for std in range(1,5)]

#rectangular box plot
plt.boxplot(data,vert=True,patch_artist=True);  
```

```python
data = [np.random.normal(0,std,100) for std in range(1,5)]

#rectangular box plot
plt.boxplot(data,vert=False,patch_artist=False);  
```

# Pie Chart

```python
labels = 'Python','C++','c','Javascript'
sizes = [110,80,90,120]
colors= ['gold','yellowgreen','lightcoral','lightskyblue']
explode = (0.1,0,0,0) #explode 1st slice

#plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%',shadow=True)
plt.axis('equal')
```
